```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from shapely.geometry import Polygon
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
states = gpd.read_file('us_eco_l3.shp')
states.head()
```

```python
def make_point(row):
    return Point(row.long, row.lat)
points = df2.apply(make_point, axis=1)
df2 = gpd.GeoDataFrame(df2, geometry=points)
df2.crs = {'init': 'epsg:4326'}
df2.head()
```

```python
states = states.to_crs(df2.crs)
states.head()
```

```python
ax = states.plot(linewidth=0.25, edgecolor='white', color='lightgrey')
df2.plot(markersize=2, alpha=0.25, color='red', ax=ax)
ax.axis('off')
```

```python
states_df = gpd.sjoin(df2, states, how='left', op='within')
states_df.head()
```

```python
ax = states.to_crs({'proj': 'merc'}).plot(color='#cccccc', edgecolor='white', figsize=(7,7))
df2.to_crs({'proj': 'merc'}).plot(ax=ax, color='red', markersize=3, alpha=0.5)
ax.axis('off')
```
