## Our data

We're going to start off with a **list of countries and their per-capita GDP in 1970 and 2015**.

This is **long format data**, so every row is an observation.

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv("country_gdp.csv")

# We're going to also pull out a subset for some of the longer graphs
wanted = ['Japan', 'Luxembourg', 'Greece', 'Iceland', 'Brazil', 'Gabon', 'India', 'United States', 'China']
subset_df = df[df.country.isin(wanted)]

df.head()
```

If you've forgotten, **wide format** looks something like this...

```python
df.pivot(index='country', columns='year', values='gdp').head()
```

## Plotting with pandas

Let's say we're interested in seeing 2015's GDP on a graph. We'll filter for 2015, then do a `.plot`. Nothing crazy.

```python
# Do this on the subset so we don't take up so much space
subset_2015 = subset_df[subset_df.year == 2015]

subset_2015.plot(y='gdp', x='country', kind='barh')
```

```python
# If we're using all countries and need it to be bigger, let's uncomment this one
#df_2015.plot(y='gdp', x='country', kind='barh', figsize=(10, 30))
```

## Plotting with Altair

### Importing

We'll start by just importing `Chart`. Eventually we'll import a lot more, which I've put below commented out.

```python
from altair import Chart
# from altair import Chart, X, Y, Axis, SortField, LayeredChart
```

When using Altair/Vega, we create a `Chart`, feeding it the data.

### Chosing a mark type

Then we decide what kind of [mark](https://altair-viz.github.io/documentation/marks.html) we're going to use to represent our data points. I think a **bar** would make sense, since we just made a bar graph.

```python
Chart(subset_2015).mark_bar()
```

### Encoding data attributes as visual attributes

But just saying "Hey, we're going to use a bar!" doesn't really *mean* anything. We need to say hey - we want the `x` and `y` axis to mean something for this bar. We want to **encode some of those data attributes as visual attributes**.

```python
Chart(subset_2015).mark_bar().encode(
    x='gdp',
    y='country'
)
```

### Color

What continent are those on? I don't remember! Let's **encode continent by color**.

```python
Chart(subset_2015).mark_bar().encode(
    x='gdp',
    y='country',
    color='continent'
)
```

## Changing our mark

There are plenty more [marks available](https://altair-viz.github.io/documentation/marks.html)! How about circles or tick marks?

```python
Chart(subset_2015).mark_circle().encode(
    x='gdp',
    y='country',
    color='continent'
)
```

```python
Chart(subset_2015).mark_tick().encode(
    x='gdp',
    y='country',
    color='continent'
)
```

### What if we wanted to squash them all together?

Don't think of it as combining them, but rather, **we're only encoding the GDP, not the country.**

```python
Chart(subset_2015).mark_tick().encode(
    x='gdp'
)
```

This would look more interesting with **all of our data**.

```python
df_2015 = df[df.year == 2015]

Chart(df_2015).mark_tick().encode(
    x='gdp'
)
```

### We could also break them down by continent 

```python
Chart(df_2015).mark_tick().encode(
    x='gdp',
    y='continent'
)
```

### ...while still coloring them

```python
Chart(df_2015).mark_tick().encode(
    x='gdp',
    y='continent',
    color='continent'
)
```

### Or maybe by year, instead?

This one actually doesn't work immediately. We'll need to say, **hey, treat year like a category - like a NOMINAL variable.**

```python
# We'll use ALL of the data this time, not df_2015
Chart(df).mark_tick().encode(
    x='gdp',
    y='year'
)
```

```python
# Treat year like a nominal variable, please!
Chart(df).mark_tick().encode(
    x='gdp',
    y='year:N'
)
```

### And of course, we can always still use color!

Not that this looks interesting, but hey, it's possible.

```python
Chart(df).mark_tick().encode(
    x='gdp',
    y='year:N',
    color='continent'
)
```

## Small multiples

Small multiples we'll talk about *later*, but instead of breaking it down by one category - year - we can break it down by *multiple categories*, like both year and continent.

We do this by establishing that the **row** is the continent, and then inside of each row, the **y** is year and the **x** is gdp.

```python
Chart(df).mark_tick().encode(
    x='gdp',
    y='year:N',
    row='continent',
    color='continent'
)
```

...and since we're just playing with variables, we could also easily flip it around and have the year be encoded in the row.

```python
Chart(df).mark_tick().encode(
    x='gdp',
    y='continent',
    row='year:N',
    color='continent'
)
```

## Encoding more categorical variables

The x or y axis is usually something **quantitative**, but it doesn't *have* to be. We could also make them something categorical, and then the intersection is where we put the quantitative value.

```python
Chart(df).mark_circle().encode(
    size='median(gdp)',
    x='continent',
    y='year:N'
)
```

## Using operations like `min`, `max`, `sum`, etc.

We can also do operations, like `min` and `max`. Remember out bar chart? **Let's make it more interesting.**

We'll have the **start of the bar** encode the minimum GDP that country had, and we'll make the **end of the bar** encode the maximum GDP that country had.

```python
Chart(subset_df).mark_bar().encode(
    x='min(gdp)',
    x2='max(gdp)',
    y='country'
)
```

Kind of unattractive! Sorting always makes everything nicer: let's **add a `SortField`** tell Altair/Vega how to sort our data.

## Creating a custom axis/sorting/ordering an axis

This is a little more complicated than I'd like it to be, but that's life! We need to create a custom `Y` axis where we tell it the variable we're encoding, and then give it a `SortField` to tell it how to order our data.

```python
from altair import Y, SortField

y_axis = Y('country', sort=SortField(field='gdp', op='max', order='descending'))

Chart(subset_df).mark_bar().encode(
    x='min(gdp)',
    x2='max(gdp)',
    y=y_axis

)
```

Let's throw in some color while we're at it! And hey, let's see **every country** for a second instead of just the subset.

```python
y_axis = Y('country', sort=SortField(field='gdp', op='max', order='descending'))

Chart(df).mark_bar().encode(
    x='min(gdp)',
    x2='max(gdp)',
    y=y_axis,
    color='continent'
)
```

## Layering graphs

While having a 'min' and a 'max' got us pretty far, we might want a graph with **different aspects of the data encoded in different marks**. aka "I only have one Albania but I want Albania to have multiple dots or multiple lines."

We'll start by encoding each datapoint as a circle, positioned according to its country and GDP.

```python
Chart(subset_df).mark_circle().encode(
    x='gdp',
    y='country',
    color='year'
)
```

Again, we might want to tell it **year is a nominal variable**, not a quantitative one (we could also say it's **ordinal** instead, give `:O` a try!).

**These are our circles.** They look nice. They'd look nicer with a line connecting them, though!

```python
Chart(subset_df).mark_circle().encode(
    x='gdp',
    y='country',
    color='year:N'
)
```

**These are our lines.** They look nice. They'd look nicer with dots on each end, though!

```python
Chart(subset_df).mark_rule().encode(
    x='min(gdp)',
    x2='max(gdp)',
    y='country'
)
```

### Building a LayeredChart

A **LayeredChart** is just charts on top of one another. It's that easy! The only difference is you feed your data to `LayeredChart`, not to the individual charts themselves.

In this case we're taking the two charts we made before and putting them on top of one another.

```python
from altair import LayeredChart

LayeredChart(subset_df, layers=[
    Chart().mark_rule().encode(
        x='min(gdp)',
        x2='max(gdp)',
        y='country'
    ),
    Chart().mark_circle().encode(
        x='gdp',
        y='country',
        color='year:N'
    )
])
```

### We probably want to sort them.

Let's sort by maximum GDP so we have a nice smooth right edge flowing down.

```python
y_axis = Y('country', sort=SortField(field='gdp', op='max', order='descending'))

LayeredChart(subset_df, layers=[
    Chart().mark_rule().encode(
        x='min(gdp)',
        x2='max(gdp)',
        y=y_axis
    ),
    Chart().mark_circle().encode(
        x='gdp',
        y=y_axis,
        color='year:N'
    )
])
```

Again, let's try it with **everything now**

```python
y_axis = Y('country', sort=SortField(field='gdp', op='max', order='descending'))

LayeredChart(df, layers=[
    Chart().mark_rule().encode(
        x='min(gdp)',
        x2='max(gdp)',
        y=y_axis
    ),
    Chart().mark_circle().encode(
        x='gdp',
        y=y_axis,
        color='year:N'
    )
])
```

## Configuring your marks

You can also [configure marks](https://altair-viz.github.io/API.html#altair.MarkConfig) to give them default colors, sizes, etc etc etc.

```python
y_axis = Y('country', sort=SortField(field='gdp', op='max', order='descending'))

LayeredChart(df, layers=[
    Chart().mark_bar().configure_mark(color='lightgrey', barSize=10, opacity=0.3).encode(
        x='min(gdp)',
        x2='max(gdp)',
        y=y_axis
    ),
    Chart().mark_tick().configure_mark(color='red', tickSize=10, tickThickness=3).encode(
        x='gdp',
        y=y_axis
    )
])
```
